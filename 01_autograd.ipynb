Calculating simple neural network

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
loss = F.binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

```python
w.grad.zero_()
b.grad.zero_()
```
