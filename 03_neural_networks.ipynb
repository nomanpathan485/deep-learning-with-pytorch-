```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([
    [1.0],
    [2.0],
    [3.0],
    [4.0]
])

y = torch.tensor([
    [2.0],
    [4.0],
    [6.0],
    [8.0]
])
```

```python
print(X.shape)
print(y.shape)
```

```python
model = nn.Linear(1,1)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01
)
```

```python
for epoch in range(1000):
    prediction = model(X)
    loss = criterion(prediction,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 100 ==0:
        print(loss.item())
```

```python
with torch.no_grad():

    test = torch.tensor([[5.0]])

    prediction = model(test)

    print(prediction)
```
